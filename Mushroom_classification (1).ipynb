**Using for loop**

```python
vals = "bruises"
backup = "odor"

for col in df.columns[:-1]:
    if col == vals:
        tmp = vals
        vals = backup
        backup = tmp

    agg = df.pivot_table(index="class", columns=col, values=vals, aggfunc=len)
    agg.fillna(0, inplace=True)

    print(f"Relation between {col} & Class")

    display(agg)

    cap_surface_e = agg.iloc[0]
    cap_surface_p = agg.iloc[1]
    cap_surface_v = agg.columns

    plt.figure(figsize=(8, 3))
    plt.subplot(1, 2, 1)
    sns.barplot(x=cap_surface_v, y=cap_surface_e)
    plt.subplot(1, 2, 2)
    sns.barplot(x=cap_surface_v, y=cap_surface_p)
    plt.subplots_adjust(hspace=.5, wspace=.5)
    plt.show()
    print("====================================================")
```

# Data Split

```python
from sqlalchemy import True_
from sklearn.model_selection import train_test_split
X,y=df.iloc[:,:-1],df.iloc[:,-1:]
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2, shuffle=True ,random_state=42 )
X_train, X_val, y_trrain, y_val = train_test_split(X,y,test_size=0.2, shuffle=True ,random_state=42 )
```

# Encoding

```python
df.nunique()
```

**Encoding the target**

```python
from sklearn.preprocessing import LabelEncoder
y_encoder=LabelEncoder()
y_train=y_encoder.fit_transform(y_train)
y_test=y_encoder.fit_transform(y_test)
y_val=y_encoder.fit_transform(y_val)
```

**Encoding the features**

```python
meta_data = pd.DataFrame({"Dtypes": X.dtypes, "Num_Uniq": X.nunique()})
meta_data
```

```python
#2types of encoder for features based on uniqe values
cols_to_bin_encode = meta_data[meta_data.Num_Uniq >= 6].index.tolist()
cols_to_1hot_encode = meta_data[meta_data.Num_Uniq < 6].index.tolist()
```

```python
#1hot for less than 6 unique
X_train = pd.get_dummies(X_train, columns=cols_to_1hot_encode)
X_val = pd.get_dummies(X_val, columns=cols_to_1hot_encode)
X_test = pd.get_dummies(X_test, columns=cols_to_1hot_encode)
```

```python
#binary for more than 6 uniqe
!pip install category_encoders

from category_encoders import BinaryEncoder
encoder = BinaryEncoder(cols_to_bin_encode)
X_train = encoder.fit_transform(X_train)
X_val = encoder.transform(X_val)
X_test = encoder.transform(X_test)
```
